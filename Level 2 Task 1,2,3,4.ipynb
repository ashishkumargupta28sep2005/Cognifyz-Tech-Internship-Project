Task: Geographic Analysis

Plot the locations of restaurants on a
map using longitude and latitude
coordinates.

Identify any patterns or clusters of
restaurants in specific areas.

```python
import folium as flm
```

```python
import pandas as pd
from folium.plugins import MarkerCluster
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
import numpy as np
```

```python
df=pd.read_csv('Dataset.csv')
```

```python
df
```

```python
df = df.dropna(subset=['Longitude', 'Latitude'])
```

```python
map_center = [df['Latitude'].mean(), df['Longitude'].mean()]
map_center
```

```python
restaurant_map = flm.Map(location=map_center, zoom_start=10)
restaurant_map
```

```python
marker_cluster = MarkerCluster().add_to(restaurant_map)
marker_cluster
```

```python
for _, row in df.iterrows():
    flm.Marker([row['Latitude'], row['Longitude']], popup=row['Restaurant Name']).add_to(marker_cluster)
```

```python
restaurant_map.save("restaurant_map.html")
print("Map saved as 'restaurant_map.html'. Open this file in a browser to view.")
```

##### -----Clustering Analysis----

```python
coords = df[['Latitude', 'Longitude']].values
coords
```

```python
db = DBSCAN(eps=0.01, min_samples=5, metric='haversine').fit(np.radians(coords))
df['Cluster'] = db.labels_
df['Cluster']
```

```python
cluster_counts = df['Cluster'].value_counts()
print("\nCluster Analysis:")
print(cluster_counts)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df['Longitude'], df['Latitude'], c=df['Cluster'], cmap='viridis', alpha=0.5)
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("Geographic Clustering of Restaurants")
plt.colorbar(label="Cluster ID")
plt.show()
```

### Level 2 Task 4

Task: Restaurant Chains

Identify if there are any restaurant chains
present in the dataset.

Analyze the ratings and popularity of
different restaurant chains.

```python
chain_counts = df['Restaurant Name'].value_counts()
print(chain_counts)
print('-'*40)

restaurant_chains = chain_counts[chain_counts > 1]
print(restaurant_chains)
```

```python
print("Top Restaurant Chains by Outlet Count:")
print(restaurant_chains.head(10))
```

```python
chain_analysis = df[df['Restaurant Name'].isin(restaurant_chains.index)].groupby('Restaurant Name').agg(
    outlet_count=('Restaurant ID', 'count'),
    avg_rating=('Aggregate rating', 'mean'),
    avg_votes=('Votes', 'mean')
).sort_values(by='outlet_count', ascending=False)

print("\nTop 10 Restaurant Chains with Ratings and Votes:")
print(chain_analysis.head(10))
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x=chain_analysis.head(10).index, y=chain_analysis['outlet_count'].head(10), palette="magma")
plt.xlabel("Restaurant Chain")
plt.ylabel("Number of Outlets")
plt.title("Top 10 Restaurant Chains by Outlet Count")
plt.xticks(rotation=45)
plt.show()
```

```python
top_rated_chains = chain_analysis.sort_values(by='avg_rating', ascending=False).head(10)
plt.figure(figsize=(10,5))
sns.barplot(x=top_rated_chains.index, y=top_rated_chains['avg_rating'], palette="viridis")
plt.xlabel("Restaurant Chain")
plt.ylabel("Average Rating")
plt.title("Top 10 Highest Rated Restaurant Chains")
plt.xticks(rotation=45)
plt.ylim(0, 5)  # Rating scale (0-5)
plt.show()
```

```python
top_popular_chains = chain_analysis.sort_values(by='avg_votes', ascending=False).head(10)
plt.figure(figsize=(10,5))
sns.barplot(x=top_popular_chains.index, y=top_popular_chains['avg_votes'], palette="coolwarm")
plt.xlabel("Restaurant Chain")
plt.ylabel("Average Votes")
plt.title("Top 10 Most Popular Restaurant Chains (by Votes)")
plt.xticks(rotation=45)
plt.show()
```
